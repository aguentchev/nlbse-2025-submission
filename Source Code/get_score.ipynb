```python
import pandas as pd
```

```python
def get_all_data_and_compare(test_csvs, baseline_path):
    combined_df = pd.concat([pd.read_csv(file) for file in test_csvs], ignore_index=True).drop(["accuracy", "Unnamed: 0"], axis=1)
    baseline_df = pd.read_csv(baseline_path).drop(["index"], axis=1)
    print(combined_df.columns)
    print(baseline_df.columns)
    if not all(combined_df.columns == baseline_df.columns):
        raise ValueError("Dataframes not the same")
    numeric_columns = ["precision", "recall", "f1"]
    for metric in numeric_columns:
        combined_df[metric] -= baseline_df[metric]
    return combined_df

def sort_highest_per_language(combined_df: pd.DataFrame):
    sorted_df = combined_df.sort_values(["lan", "f1"], ascending=[True, False])
    highest_per_language = sorted_df.groupby("lan")
    return highest_per_language
```

```python
df = get_all_data_and_compare(["./results/java_final_test_result.csv", "./results/pharo_final_test_result.csv", "./results/python_final_test_result.csv"], "baseline_results_summary.csv")
df
```
